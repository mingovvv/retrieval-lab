# Model Comparison

Use this notebook to compare providers, models, and retrieval settings.

```python
from _bootstrap import setup_project_path

setup_project_path()
```

```python
from embedding_retrieval.config import RetrievalConfig
from embedding_retrieval.factory import create_retrieval_pipeline

EXPERIMENTS = [
    {"embedding_provider": "fake", "embedding_model": "fake", "chunk_size": 200},
    {"embedding_provider": "fake", "embedding_model": "fake", "chunk_size": 400},
]

queries = ["retrieval workflows", "local embedding model"]
texts = [
    "LangChain supports retrieval workflows.",
    "Ollama runs local embedding models.",
    "FAISS is useful for dense vector similarity search.",
]
```

```python
comparison_rows = []
base_config = RetrievalConfig.from_env()

for spec in EXPERIMENTS:
    config = base_config.with_overrides(
        embedding_provider=spec["embedding_provider"],
        embedding_model=spec["embedding_model"],
        chunk_size=spec["chunk_size"],
        chunk_overlap=0,
        top_k=2,
    )
    ingest_service, search_service = create_retrieval_pipeline(config)
    ingest_service.add_texts(texts)
    for query in queries:
        results = search_service.search(query)
        comparison_rows.append({
            "provider": spec["embedding_provider"],
            "model": spec["embedding_model"],
            "chunk_size": spec["chunk_size"],
            "query": query,
            "top_result": results[0].document.page_content if results else None,
            "score": results[0].score if results else None,
        })

comparison_rows
```
