# Dataset Evaluation

Use a fixed query set to inspect retrieval quality against your own corpus.

```python
from _bootstrap import setup_project_path

setup_project_path()
```

```python
from embedding_retrieval.config import RetrievalConfig
from embedding_retrieval.factory import create_retrieval_pipeline

config = RetrievalConfig.from_env().with_overrides(chunk_size=300, chunk_overlap=30, top_k=3)
ingest_service, search_service = create_retrieval_pipeline(config)
```

```python
corpus = [
    "Document one about semantic search and retrieval.",
    "Document two about local model serving with Ollama.",
    "Document three about vector indexing with FAISS.",
]
queries = [
    {"query": "semantic retrieval", "expected_keyword": "retrieval"},
    {"query": "local model", "expected_keyword": "Ollama"},
]

ingest_service.add_texts(corpus)
```

```python
evaluation_rows = []

for item in queries:
    results = search_service.search(item["query"], top_k=1)
    top_text = results[0].document.page_content if results else ""
    evaluation_rows.append({
        "query": item["query"],
        "top_result": top_text,
        "matched_expected": item["expected_keyword"].lower() in top_text.lower(),
    })

evaluation_rows
```
